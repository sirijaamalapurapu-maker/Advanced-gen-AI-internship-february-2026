Problem 1: Unique Words in a Sentence


```python
sentence=input("Enter sentence: ")

words =sentence.split()
unique_words=[]

for word in words:
    if word not in unique_words:
        unique_words.append(word)

print("Unique words:", set(unique_words))
print("Number of unique words: ",len (unique_words))
```

Problem 2: Highest Salary from Employee Data


```python
employees = {"Ravi": 75000,"Anita": 68000,"Kiran": 72000}

highest_salary=0
highest_employee=""

for name in employees:
  if employees[name] > highest_salary:
    highest_salary= employees[name]
    highest_employee=name

print( "highest_salary: ",highest_employee,"-",highest_salary)
```

Problem 5: Calculate Attendance Percentage


```python
attendance = ["P", "P", "A", "P", "P"]

present_count = 0
total_days = len(attendance)

for day in attendance:
    if day == "P":
        present_count += 1

percentage = (present_count / total_days) * 100

print("Attendance Percentage:", percentage)
```

Problem 7: Count Character Frequency



```python
text = "python"

char_count = {}

for char in text:
    if char in char_count:
        char_count[char] += 1
    else:
        char_count[char] = 1

print(char_count)
```
