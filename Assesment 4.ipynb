Problem 1: Mobile Recharge Validation System


```python
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]

    if amount < 50:
        print("Recharge amount must be at least ₹50.")
        return False
    elif amount not in valid_plans:
        print("Invalid plan selected.")
        print("Available plans:", valid_plans)
        return False
    else:
        print("Recharge successful for ₹", amount)
        return True

while True:
    try:
        user_amount = int(input("Enter recharge amount: "))

        if validate_recharge(user_amount):
            break

        print("Please try again.\n")

    except ValueError:
        print("Please enter a valid numeric amount.\n")
```

Inventory Reorder Alert System


```python
def check_stock(products):
    for product, quantity in products.items():
        if quantity < 15:
            print(f"Reorder Alert: {product} (Stock: {quantity})")
        else:
            print(f"Stock perfect: {product} (Stock: {quantity})")


inventory = {
    "Rice": 20,
    "Sugar": 10,
    "Oil": 25,
    "Salt": 8,
    "Flour": 18
}

check_stock(inventory)
```

Student Result Processing System

```python
def process_result(marks):
    total = 0

    # Calculate total
    for mark in marks:
        total += mark

    average = total / len(marks)

    if average >= 50:
        return f"Average: {average:.2f} - Pass"
    else:
        return f"Average: {average:.2f} - Fail"


student_marks = [60, 55, 70, 40, 65]

result = process_result(student_marks)
print(result)
```

Cab Fare Estimator with Retry Option

```python
def calculate_fare(distance, peak):
    base_fare = 50
    rate_per_km = 12

    fare = base_fare + (distance * rate_per_km)

    if peak.lower() == "yes":
        fare += fare * 0.25   # 25% extra charge

    return fare


# Retry system using while loop
while True:
    try:
        distance = float(input("Enter distance in km: "))
        peak = input("Is it peak hour? (yes/no): ")

        total_fare = calculate_fare(distance, peak)

        print(f"Total Cab Fare: ₹{total_fare:.2f}")

        choice = input("Do you want to calculate again? (yes/no): ")
        if choice.lower() != "yes":
            print("Thank you for using the Cab Fare Estimator!")
            break

    except ValueError:
        print("Please enter valid numeric distance.\n")
```

Employee Attendance Eligibility Checker

```python
def check_attendance(attendance_list):
    total_days = len(attendance_list)
    present_days = 0

    for status in attendance_list:
        if status.upper() == "P":
            present_days += 1

    percentage = (present_days / total_days) * 100

    if percentage >= 75:
        return f"Attendance: {percentage:.2f}% - Eligible"
    else:
        return f"Attendance: {percentage:.2f}% - Not Eligible"


attendance = ["P", "A", "P", "P", "A", "P", "P", "P"]

result = check_attendance(attendance)
print(result)
```

Password Strength Checker

```python
def check_password_strength(password):
    if len(password) < 8:
        return "Weak Password: Minimum length should be 8 characters."

    has_digit = False
    has_special = False
    special_chars = "@#$"

    # Loop through each character
    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_chars:
            has_special = True

    if not has_digit:
        return "Weak Password: Must contain at least one digit."

    if not has_special:
        return "Weak Password: Must contain at least one special character (@#$)."

    return "Strong Password"


password = input("Enter password: ")
result = check_password_strength(password)
print(result)
```
